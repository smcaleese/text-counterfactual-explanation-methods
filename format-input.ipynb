```python
import pandas as pd

input_file = "sst-input"
df_input = pd.read_csv(f"./input/{input_file}.csv")
df_input.head(20)
```

```python
df_input.shape
```

```python
substitutions = {
    " .": ".", # Remove space before period
    " ,": ",", # Remove space before comma
    " 's": "'s", # Remove space before 's
    "( ": "(", # Remove space after opening parenthesis
    " )": ")", # Remove space before closing parenthesis
    "s '": "s'", # Remove space before ' in possessive
    " :": ":", # Remove space before colon
    " ;": ";", # Remove space before semi-colon
    " n't": "n't", # Remove space before n't
    " n ' t": "n't", # Remove spaces around n't
    "i": "I" # Capitalize i
}

def apply_substitutions(text, substitutions):
    for old, new in substitutions.items():
        text = text.replace(old, new)
    return text

def format_sentence(sentence):
    sentence = sentence.strip()
    sentence = apply_substitutions(sentence, substitutions)
    sentence = sentence.capitalize()
    return sentence

# Apply the reformatting function to the data
df_input['formatted_text'] = df_input['original_text'].apply(format_sentence)
df_input.head(20)
```

```python
df_input.to_csv(f"./input/sst-input-formatted.csv", index=False)
```

```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
data = {
    'Condition': ['Condition 1', 'Condition 2', 'Condition 3'],
    'Method 1': [20, 35, 30],
    'Method 2': [25, 32, 34]
}

df = pd.DataFrame(data)

# Setting the positions and width for the bars
positions = np.arange(len(df['Condition']))
width = 0.35

fig, ax = plt.subplots()

# Creating bars for each method
bars1 = ax.bar(positions - width/2, df['Method 1'], width, label='Method 1')
bars2 = ax.bar(positions + width/2, df['Method 2'], width, label='Method 2')

# Adding some labels and a title
ax.set_xlabel('Condition')
ax.set_ylabel('Values')
ax.set_title('Values by Condition and Method')
ax.set_xticks(positions)
ax.set_xticklabels(df['Condition'])
ax.legend()

# Display the plot
plt.show()
```
